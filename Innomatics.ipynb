```python
import pandas as pd
import json

# Task 1: Data Merging
def merge_csv_files(file_paths):
    """
    Merges multiple CSV files into a single DataFrame.
    Args:
        file_paths: List of file paths to merge.
    Returns:
        Merged DataFrame.
    """
    dataframes = []
    for file_path in file_paths:
        try:
            # Read each CSV file and append to the list
            df = pd.read_csv(file_path, encoding='ISO-8859-1')
            dataframes.append(df)
        except Exception as e:
            print(f"Error reading file {file_path}: {e}")
    # Concatenate all DataFrames
    merged_df = pd.concat(dataframes, ignore_index=True)
    return merged_df


# Specify the paths to the CSV files
file_paths = [
    "C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Bilekahalli.csv",
    "C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\BTM 2nd Stage.csv",
    "C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Electronic City.csv",
    "C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Hebbal.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Horamavu.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Hosakerehalli.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Hulimavu.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Indiranagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Jayanagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kalyan Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kammanahalli.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Krishnarajapura.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kumaraswamy Layout.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Lingarajapuram.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Mahadevapura.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Marathahalli.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Nagarbhavi.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Padmanabhanagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Rajaji Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Ramamurthy Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\RR Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\RT Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Subramanyapura.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Uttarahalli Hobli.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Vijaya Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Yelahanka.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Akshayanagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Arekere.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Basavanagudi.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Basaveshwar Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Begur.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Bommanahalli.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Brookefield.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\BTM_Layout.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Chikkalasandra.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Doddanekundi.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Ejipura.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Electronics City Phase 1.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Hongasandra.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Hoodi.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\HSR_Layout.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\JP Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\K.R Puram.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kadugodi.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kaggadasapura.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kengeri Satellite Town.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kengeri.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Konanakunte.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Koramangala.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Mathikere.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Munnekollal.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Raja Rajeshwari Nagar.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Singasandra.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Thanisandra.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Varthur.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Vidyaranyapura.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Whitefield.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Gottigere.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\HBR Layout.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Kasavanahalli.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Sunkadakatte.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Yelahanka New Town.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Banashankari.csv",
"C:\\Users\\Bhavani\\Downloads\\Property_data-20250109T054843Z-001\\Property_data\\Bellandur.csv"
]

# Merge the property CSV files
merged_property_data = merge_csv_files(file_paths)

# Save the merged DataFrame
merged_property_data.to_csv("merged_property_data.csv", index=False)
print("Merged DataFrame shape:", merged_property_data.shape)

# Task 2: Data Cleaning
def clean_photo_urls(photo_data):
    """
    Cleans the photo_urls column to extract the number of photos for each property.
    Args:
        photo_data: DataFrame containing the photo_urls column.
    Returns:
        Updated DataFrame with photo_count feature.
    """
    def extract_photo_count(photo_urls):
        if pd.isna(photo_urls):
            return 0
        try:
            photo_list = json.loads(photo_urls.replace("'", '"'))  # Handle corrupted JSON-like strings
            return len(photo_list) if isinstance(photo_list, list) else 0
        except json.JSONDecodeError:
            return 0

    photo_data["photo_count"] = photo_data["photo_urls"].apply(extract_photo_count)
    return photo_data


# Load and clean the photo dataset
property_photos_data = pd.read_csv("C:\\Users\\Bhavani\\Downloads\\property_photos.tsv", sep='\t')
cleaned_photos_data = clean_photo_urls(property_photos_data)

# Task 3: Feature Engineering
# Load property interactions data
property_interactions_data = pd.read_csv("C:\\Users\\Bhavani\\Downloads\\property_interactions.csv")

# Calculate total interactions for each property
interactions_summary = property_interactions_data.groupby("property_id").size().reset_index(name="total_interactions")

# Merge photo count and interaction data with the main property dataset
final_data = merged_property_data.merge(
    cleaned_photos_data[["property_id", "photo_count"]], on="property_id", how="left"
)
final_data = final_data.merge(interactions_summary, on="property_id", how="left")

# Fill missing values for new features
final_data["photo_count"].fillna(0, inplace=True)
final_data["total_interactions"].fillna(0, inplace=True)

# Task 4: EDA Preparation
# Examples of EDA questions to analyze
eda_results = {
    "Top 5 Properties by Total Interactions": final_data.nlargest(5, "total_interactions")[["property_id", "total_interactions"]],
    "Average Rent by Property Type": final_data.groupby("type")["rent"].mean(),
    "Properties with Gym and Swimming Pool": final_data[(final_data["gym"] == 1) & (final_data["swimming_pool"] == 1)],
    "Properties by Furnishing Status": final_data["furnishing"].value_counts(),
}

# Save the final DataFrame
final_data.to_csv("final_property_data.csv", index=False)

# Display EDA results
for question, result in eda_results.items():
    print(f"\n{question}:\n{result}")
```
